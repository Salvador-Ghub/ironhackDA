```python
nombre = input("Ingresa tu nombre: ")
edad = int(input("Ingresa tu edad: "))

def futuro10(edad_actual):
    return edad_actual + 10

if edad < 18:
    print('Eres menor de edad.')
else:
    print('Eres mayor de edad.')

print(f"Hola {nombre}, en 10 años tendrás {futuro10(edad)} años.")
```

```python
def grupo_edad(edad):
    if edad < 12:
        return("Niño")
    elif 12 <= edad <= 17:
        return("Adolescente")
    elif 18 <= edad <= 64:
        return("Adulto")
    else:  # edad >= 65
        return("Adulto mayor")

edad = int(input("Ingresa tu edad: "))

print(grupo_edad(edad))
```

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.array([1, 2, 3, 4, 5])

cuadrados_array = array ** 2

plt.figure(figsize=(8, 5))
plt.plot(array, cuadrados_array, marker='X')
plt.ylabel("Cuadrados")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

data = {
    "Nombre": ["Carlos", "Miguel", "Laia", "Víctor", "Oriol"],
    "Edad": [13, 19, 33, 21, 21],
    "Calificación": [4, 5, 9, 8, 4]
}

df = pd.DataFrame(data)

df["Aprobado"] = df["Calificación"] >= 60

promedio = df["Calificación"].mean()

print(df)
print(f"\nCalificación promedio: {promedio:.2f}")
```
