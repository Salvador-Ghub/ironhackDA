## 6. Escribe una funciÃ³n que devuelva el factorial de un nÃºmero.

```python
def factorial(n):
#tu codigo aquÃ­
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact

print(factorial(6))
```

```python
# FÃ³rmula factorial
# n! = n * (n - 1) *...*1

# Este cÃ³digo define una funciÃ³n llamada "factorial" que toma una entrada "n". La funciÃ³n utiliza un bucle for para iterar a travÃ©s del rango de nÃºmeros
# desde 1 hasta n+1. Para cada nÃºmero en ese rango, multiplica el valor actual de x por el nÃºmero en el rango. Al final del bucle,
# la funciÃ³n devuelve el valor final de x, que serÃ¡ el factorial del nÃºmero de entrada "n".

# El factorial de un entero positivo n es el producto de todos los enteros positivos menores o iguales a n.
# Por ejemplo, el factorial de 6 (escrito "6!") es 6 * 5 * 4 * 3 * 2 * 1 = 720.

# AsÃ­ que esta funciÃ³n toma una entrada de cualquier entero positivo y devuelve el factorial de ese nÃºmero.
```

```python
def sort_alpha(string):
        
    def sep_palabras(string):
        word=""
        word_list=[]
        pos_list=0
        for letter in string:
            #print(letter)
            if letter==",":
                word_list.append(word)
                pos_list+=1
                word=""
            else:
                word+=letter
                #print(word)
        word_list.append(word)
        return word_list

    def ordenar_palabras(lista):
        return sorted(lista)

    def junt_palabras(lista):
        string=""
        for palabra in lista:
            for caracter in palabra:
                string+=caracter
            string+=","
        return string[:-1]

    return junt_palabras(ordenar_palabras(sep_palabras(string)))

string="Ahora,Desterrado,Caballero,Buen,FrancÃ©s,Estrellado"
print(sort_alpha(string))
```
