```python
# Dataset quality check function
def quality_checks(df):
    x = df
    # Quality analysis of the database
    print("TOTAL NUMBER OF ROWS: \n",len(x),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -\n")
    print("NUMBER OF UNIQUE VALUES:\n",x.nunique(),"\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("NUMBER OF UNIQUE VALUES DIVIDED BY TOTAL VALUES:\n",round((x.nunique()/len(x)*100),2),"%\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print("% OF NAN VALUES IS: \n",round(100*(x.isnull().sum() / len(x)),2),"%\n")
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - \n")
    print(x.describe())

quality_checks(df_Train_csv_original)
```

Considering that no numerical value appears to be an outlier according to the analysis, the filling of NAN values will proceed.
According to the data analysis, we have a small percentage of NAN values. These will be replaced with the modes and means of each column to avoid losing relevant information.

```python
# Print value counts for each Categorical column
def categorical_value_counts(df):
    df_filled = df.copy()  # Create a copy to avoid undesired modifications.
    for column in df_filled.columns:
        if df_filled[column].dtype in ['float64', 'int64']:  # It is numeric
            pass
        else:  # It is categorical
            print(df_filled[column].value_counts())
            print("\n")
    return

categorical_value_counts(df_Train_csv_original)
```

```python
# Function to fill missing values

def fill_missing_values(df):
    df_filled = df.copy()  # Create a copy to avoid undesired modifications.
    for column in df_filled.columns:
        if df_filled[column].dtype in ['float64', 'int64']:  # It is numeric
            df_filled[column] = df_filled[column].fillna(df_filled[column].mean())
        else:  # It is categorical
            mode_value = df_filled[column].mode()[0] if not df_filled[column].mode().empty else np.nan
            df_filled[column] = df_filled[column].fillna(mode_value)
    return df_filled


# To fill missing values
df_Train_csv_original_filled = fill_missing_values(df_Train_csv_original)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Crear el boxplot por cada 'Unit'
plt.figure(figsize=(8,6))  # Tama√±o de la figura
sns.boxplot(x='Decision_skill_possess', y='Attrition_rate', data=df_Train_csv_original_filled)

# Mostrar el gr√°fico
plt.title('Boxplot de Attrition Rate por Unidad')
plt.show()

def calculate_outliers_by_group(df, group_column, target_column):
    # Diccionario para almacenar los outliers por cada grupo
    outliers_by_group = {}
    
    # Agrupar el DataFrame por la columna de grupo (por ejemplo, 'Unit')
    grouped = df.groupby(group_column)
    
    # Iterar sobre cada grupo
    for group, group_df in grouped:
        # Calcular los cuartiles y el rango intercuartil (IQR)
        Q1 = group_df[target_column].quantile(0.25)
        Q3 = group_df[target_column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Definir los l√≠mites para identificar outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identificar los outliers
        outliers = group_df[(group_df[target_column] < lower_bound) | (group_df[target_column] > upper_bound)]
        
        # Almacenar los outliers de cada grupo en el diccionario
        outliers_by_group[group] = outliers.shape[0]
    
    return outliers_by_group

# Llamar a la funci√≥n para obtener los outliers por cada grupo
outliers = calculate_outliers_by_group(df_Train_csv_original_filled, 'Decision_skill_possess', 'Attrition_rate')

# Mostrar los resultados
print(outliers)
```
